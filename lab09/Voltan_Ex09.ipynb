```python
# LIBRARIES
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# DATA FILES
pos_c = np.loadtxt("data/lab09/circle_pos.out");
best_c = np.loadtxt("data/lab09/circle_best.out", dtype = "int");
best_c = np.append(best_c, best_c[0]);
mean_c = np.loadtxt("data/lab09/circle_mean.out");
mean_s = np.loadtxt("data/lab09/square_mean.out");
pos_s = np.loadtxt("data/lab09/square_pos.out");
best_s = np.loadtxt("data/lab09/square_best.out", dtype = "int");
best_s = np.append(best_s, best_s[0]);
```

```python
x = np.arange(mean_c.size // 2);
plt.figure(figsize=(15,4))
plt.suptitle("TSP on a circle", fontsize = "xx-large")
plt.subplot(1, 2, 1)
plt.title("L1 Cost decay")
plt.plot(x, mean_c[:,1], label="Best Path");
plt.plot(x, mean_c[:,0], label="Average Best Half");
plt.xlabel(r"# Generations")
plt.ylabel(r"Cost")
plt.legend()
plt.grid(True)
plt.subplot(1, 2, 2)
plt.title("Best path")
plt.plot(pos_c[best_c, 0], pos_c[best_c, 1]);
plt.xlabel(r"x")
plt.ylabel(r"y")
print(f"AVERAGE cost = {mean_c[-1, 0]:.4f}")
print(f"BEST PATH cost = {mean_c[-1, 1]:.4f}")
plt.plot();
```

```python
x = np.arange(mean_s.size // 2);
plt.figure(figsize=(15,4))
plt.suptitle("TSP inside a square", fontsize = "xx-large")
plt.subplot(1, 2, 1)
plt.title("L1 Cost decay")
plt.plot(x, mean_s[:,1], label="Best Path");
plt.plot(x, mean_s[:,0], label="Average Best Half");
plt.xlabel(r"# Generations")
plt.ylabel(r"Cost")
plt.legend()
plt.grid(True)
plt.subplot(1, 2, 2)
plt.title("Best path")
plt.plot(pos_s[best_s, 0], pos_s[best_s, 1]);
plt.xlabel(r"x")
plt.ylabel(r"y")
print(f"AVERAGE cost = {mean_s[-1, 0]:.4f}")
print(f"BEST PATH cost = {mean_s[-1, 1]:.4f}")
plt.plot();
```
